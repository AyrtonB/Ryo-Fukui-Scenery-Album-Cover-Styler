### Imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from PIL import Image
```

<br>

### Helper Functions

```python
def hide_spines(ax, positions=['top', 'right']):
    """
    Pass a matplotlib axis and list of positions with spines to be removed
    
    args:
        ax:          Matplotlib axis object
        positions:   Python list e.g. ['top', 'bottom']
    """
    assert isinstance(positions, list), 'Position must be passed as a list '
    
    for position in positions:
        ax.spines[position].set_visible(False)
```

<br>

### Reading in and Visualising the Data

Here the data is read in, converted to an array, flattened from 3d to 2d, and then finally transformed to grayscale

```python
rgb_2_grey = lambda rgb: np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

img_jpg = Image.open('data/bowie.jpg')

img_array = np.asarray(img_jpg.getdata())
img = img_array.reshape(img_jpg.size[1], img_jpg.size[0], 3)

img_bw = rgb_2_grey(img)
img_bw = img_bw

plt.imshow(img_bw, cmap='gray')
```

<br>

### Resizing Image

The next step is to resize the image to a resolution at which the dots will be visible. The first part of this involves determining the factors by which the row and column sizes are divisible.

```python
def print_factors(x):
    print(f'The factors of {x} are:')
    
    for i in range(1, x + 1):
        if x % i == 0:
            print(i)

            
print_factors(img_bw.shape[0])
```

<br>

Once an optimal value has been determined the image can be resized

```python
def convert_2_int(float_val):
    assert float_val%1 == 0, 'float will leave a remainder if converted to an integer'
    int_val = int(float_val)
    
    return int_val

shrink_img_arr = lambda data, rows, cols: data.reshape(rows, convert_2_int(data.shape[0]/rows), cols, convert_2_int(data.shape[1]/cols)).sum(axis=1).sum(axis=2)

new_width = 61

shrunk_img = shrink_img_arr(img_bw, new_width, new_width)
shrunk_img = shrunk_img/shrunk_img.max()

plt.imshow(shrunk_img, cmap='gray')
```

<br>

### Dot Size from Grayscale Values

The array is converted to a dataframe and unstacked so that the x and y values can be explicitly shown alongside their dot size

```python
df_bw = (pd.DataFrame(shrunk_img)
         .unstack()
         .reset_index()
         .rename(columns={
             'level_0' : 'x',
             'level_1' : 'y',
             0 : 'dot_size',
         })
        )

df_bw['y'] = df_bw['y'].max() - df_bw['y']

df_bw.head()
```

<br>

### Visualising

Everything is now ready to plot

```python
fig, ax = plt.subplots(dpi=500)

ax.scatter(df_bw['x'], df_bw['y'], s=(1-df_bw['dot_size'])*10, color='k')

ax.set_facecolor('red')
ax.set_aspect('equal')

ax.set_xlim(0, new_width-1)
ax.set_ylim(0, new_width-1)

hide_spines(ax, positions=['top', 'bottom', 'left', 'right'])
ax.set_xticks([], [])
ax.set_yticks([], [])

fig.savefig('data/bowie_stylised.png')
```
